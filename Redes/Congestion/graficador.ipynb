```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with the provided points

x1 = range(1, 7)
x2 = range(7, 11)
x3 = range(11, 17)
x4 = range(17, 21)
x5 = range(21, 25)
x = list(x1) + list(x2) + list(x3) + list(x4) + list(x5)

y = [2**(i-1) for i in x1] + [32 + i - 6 for i in x2] + [21 + i - 11 for i in x3]
y += [2**(i-17) for i in x4] + [13 + i-21 for i in x5]

z12 = [32 for i in range(1, 11)]
z3 = [21 for i in range(10,17)]
z45 = [13 for i in range(16,25)]

window_data = {'x': x, 'y': y}
window_df = pd.DataFrame(window_data)
ssthresh_df_1 = pd.DataFrame({'x': list(x1) + list(x2), 'z': z12})
ssthresh_df_2 = pd.DataFrame({'x': [10] + list(x3), 'z': z3})
ssthresh_df_3 = pd.DataFrame({'x': [16] + list(x4) + list(x5), 'z': z45})


# Plot the points
plt.figure(figsize=(16, 9))
plt.plot(ssthresh_df_1['x'], ssthresh_df_1['z'], linestyle='--', lw=2, color='tab:orange', label='ssthresh')
plt.plot(ssthresh_df_2['x'], ssthresh_df_2['z'], linestyle='--', lw=2, color='tab:orange')
plt.plot(ssthresh_df_3['x'], ssthresh_df_3['z'], linestyle='--', lw=2, color='tab:orange')
plt.plot(window_df['x'], window_df['y'], marker='o', linestyle='-', lw=2, color='tab:blue', label='Points')



# the y axis should have a 2 size

# Remove the top and right spines
ax = plt.gca()  # Get current axes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

# Add labels and title
plt.xlabel('Transmission round', fontsize = 16)
plt.xticks(range(0, 25, 2))
plt.ylabel('Window size', fontsize = 16)
plt.yticks(range(0, 40, 2))
plt.title('TCP Congestion control', fontsize = 20)
plt.grid(True)
plt.legend(fontsize = 20)
plt.show()
```
